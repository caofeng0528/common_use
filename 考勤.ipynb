```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('D:/01-1(1).xls')

dict_1 = {}
for j in data['日期'].unique():
    dict_2 = {}
    for i in data['考勤号码'].unique():
        dict_3 = {}
        name_time = data.loc[(data['考勤号码'] == i)&(data['日期'] == j)]
        if len(name_time['考勤号码']) != 0:
            if len(name_time['考勤号码']) == 1:
                if name_time['时间'].iloc[0].hour>12:
                    dict_3[i] = round((name_time['时间'].iloc[0].hour*3600+name_time['时间'].iloc[0].minute*60+name_time['时间'].iloc[0].second-9*3600)/3600,2)
                if name_time['时间'].iloc[0].hour<12:
                    dict_3[i] = round((18*3600-name_time['时间'].iloc[0].hour*3600+name_time['时间'].iloc[0].minute*60+name_time['时间'].iloc[0].second)/3600,2)
            else:
                dict_3[i] = round((max(name_time['时间']).hour*3600+max(name_time['时间']).minute*60+max(name_time['时间']).second-min(name_time['时间']).hour*3600+min(name_time['时间']).minute*60+min(name_time['时间']).second)/3600,2)
        
        else:
            continue
        dict_2[i] = dict_3
    dict_1[j] = dict_2
finall = pd.DataFrame.from_dict(dict_1,orient='index').T
finall
```
